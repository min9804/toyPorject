```python
def magic(file_path): #magic 함수지정

    import xlrd #엑셀 데이터 가져오는 모듈
    import operator #함수로서 표준 연산자
    wb = xlrd.open_workbook(file_path) #xlrd 모듈의 open_workbook 함수,  wb를 open_workbook함수의 인수로 줌,  open_workbook함수가 해당 디렉터리로 찾아가 파일을 열고 이를 wb변수에 저장한다.
    # filepath에 들어가는 경로로 들어가 파일을 열고 파일 내용을 wb 인수에 저장

    per_sh = wb.sheet_by_name('per') #per_sh는 wb에 저장된 per이라는 시트의 값을 받음

    per_dict = {} #per_dict 이라는 빈 딕셔너리 생성 
    for i in range(1, per_sh.nrows): # i 가 1(맨윗줄제외(회사명, per)) ~ per_sh.nrows(데이터가 몇 행 까지알려주는 데이터) 에 있을때
        data = per_sh.row_values(i)  #data라는 변수는 per시트의 낮은행 을 지정
        name = data[0] #딕셔너리에 key 값이 그행에 첫번째열 즉 회사명
        per = data[1] #딕셔너리에  values 값이 두번째열 즉 per
        if per > 0 :    #per 이 0보다 클때만 per_dict딕셔너리의 이름과 per 값이 일치
            per_dict[name] = per 

    sorted_per = sorted(per_dict.items(), key=operator.itemgetter(1)) #sorted(오름차순 정렬) #per_dict.items() 정렬의 대상을 키 값 쁜만 아니라 키와 그에 대응 하는 데이터 값으로 설정한 것이다. 결과값을 보면 기 값과 그에 대응하는 데이터 값이 가로로 묶여나온다. 여기서 원소가 두개의 값을 가지고 있으므로 어떤 값을 기준으로 정렬할지 정해주어야 한다. #oprator.itemgrtter(1) 두쌍중에 두번 째 값, 파이썬인덱스로 1번 값을 사용하겠다는 것이다. (파이썬 인덱스는 0번부터 메긴다.) 따라서 1번(즉 두번째) 값은 per값이므로 per값을 오름차순으로 정렬한다. 

    per_rank = {} #per_rank 라는 빈 딕셔너리 생성
    for num, firm in enumerate(sorted_per): #enumerate는 for문이 돌면서 지금이 몊 번째 인지 알려 주는 함수로 enumerate가 사용 되면 for문의 임시변수가 두 개가 된다. 변수num은 현재 for문의 순서이고, firm은 리스트의 원소이다. 
        per_rank[firm[0]] = num + 1 #firm(리스트 의 첫번재 원소) = num + 1 (1등, 2등... 순서매김)
    
    roa_sh = wb.sheet_by_name('roa') 

    roa_dict = {}
    for i in range(1, roa_sh.nrows):
        data = roa_sh.row_values(i)
        name = data[0]
        roa = data[1]
        if roa !='':  #roa 값은 -값을 가져도 되므로 데이터가 누락된 것만 걸러준다.
            roa_dict[name] = roa
    
    sorted_roa = sorted(roa_dict.items(), key=operator.itemgetter(1), reverse=True) #roa는 내림차순으로 정렬해야 하기때문에 reverse라는 옵션을 True로 세팅하면 내림차순으로 정렬된다. 기본 정렬이 오름차순이기때문에 reverse 해주면 내일 차순. 반대도 가능

    roa_rank = {}
    for num, firm in enumerate(sorted_roa):
        roa_rank[firm[0]] = num + 1

    total_rank = {} #total_rank 라는 빈 딕셔너리 생성
    for name in roa_rank.keys():  #name 이 roa_rank안 keys 값안에 있을때 
        if name in per_rank.keys(): #name 이 per_rank.keys 안에 있을때
            total_rank[name] = per_rank[name] + roa_rank[name] #taotal_rank 는 per + roa rank 이다

    sorted_total = sorted(total_rank.items(), key=operator.itemgetter(1)) #total 딕셔너리 랭크 정렬

    magic_rank = {}   #magic_rank 빈 딕셔너리 생성
    for num, firm in enumerate(sorted_total): 
        magic_rank[firm[0]] = num + 1 #magic_rank 순서 매김

    return magic_rank  #magic_rank 값 돌려줌
```
