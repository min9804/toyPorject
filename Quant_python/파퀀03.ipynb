```python
def make_fs_dataframe(firm_code): #재무데이터를 종목코드에 를 바꿔가며 가공하는 함수 만듬
    fs_url = 'https://comp.fnguide.com/SVO2/asp/SVD_Finance.asp?pGB=1&cID=&MenuYn=Y&ReportGB=D&NewMenuID=103&stkGb=701&gicode=' + firm_code #fn가이드 재무데이터 주소에 종목코드만 바꾸면 주소가 바뀌는 주소를 fs_url에 저장
    fs_page = requests.get(fs_url) #requests.get함수를 이용해 fs_url에있는 정보를 불러와 fs_page에 저장
    fs_tables = pd.read_html(fs_page.text) #fs_page.text를 pd.read_html 함수에 넘겨 fs_tables라는 데이터 프레임을 만듬
 
    temp_df = fs_tables[0] #fs_tables[0]을 temp_df에 저장
    temp_df = temp_df.set_index(temp_df.columns[0]) #IFRS(연결)이 GAAP인 회사도 있기 때문에. 인덱스로 세팅할 데이터는 첫 번째 칼럼에 있으므로 temp_df.columns[0]코드로 첫번재 칼럼명을 가져와서 set_index에 넣어줌 # temp_df에 저장
    temp_df = temp_df[temp_df.columns[:4]] # tmep_df.columns[:4]즉 4개 년도만 temp_df에 저장
    temp_df = temp_df.loc[['매출액', '영업이익', '당기순이익']] #loc함수를 이용해 원하는 정보만 남김

    temp_df2 = fs_tables[2] #[2]데이터를 temp_df2에 저장
    temp_df2 = temp_df2.set_index(temp_df2.columns[0]) #칼럼명으로 인덱스 저장
    temp_df2 = temp_df2.loc[['자산', '자본', '부채']] #원하는 것만남김

    temp_df3 = fs_tables[4]
    temp_df3 = temp_df3.set_index(temp_df3.columns[0])
    temp_df3 = temp_df3.loc[['영업활동으로인한현금흐름']]

    fs_df = pd.concat([temp_df, temp_df2, temp_df3]) #concat함수를 이용해 같은 칼럼을 공하는 데이터 프레임을 위아래로 합쳐서 fs_df에 저장
 
    return fs_df #fs_df 리턴
```

```python
#앞의 코드를 전 칼럼에 범위를 넓혀서 적용시킨 코드이다.
code = 'A005930'
fs_df = make_fs_dataframe(code)

for num, col in enumerate(fs_df.columns): #enumerate 를 이용해 for문에 순서num, 리스트의 원소col 지정 #fs_df.columns가 참일때
    temp_df = pd.DataFrame({code : fs_df[col]}) #fs_df[col]순서대로 칼럼을 가져와 새로운 데이터 프레임 만듬
    temp_df = temp_df.T #행과열을 바꿈
    temp_df.columns = [[col]*len(fs_df), temp_df.columns] #날짜 표시해주는 코드
    if num == 0:  
        total_df = temp_df # 첫 칼럼을 total_df에저장
    else:
        total_df = pd.merge(total_df, temp_df, how='outer', left_index=True, right_index=True)
        #2번째 칼럼부터 totla_df 에 merge함수를 이용해 outer(합집합)로 병합
```

```python
def change_df(firm_code, dataframe): #데이터프레임의 행과열을 바꾸는 함수 만듬 
    for num, col in enumerate(dataframe.columns):
        temp_df = pd.DataFrame({firm_code : dataframe[col]})
        temp_df = temp_df.T
        temp_df.columns = [[col]*len(dataframe), temp_df.columns]
        if num == 0:
            total_df = temp_df
        else:
            total_df = pd.merge(total_df, temp_df, how='outer', left_index=True, right_index=True)
    
    return total_df    
```

```python
firmcode_list = ['A005930', 'A005380', 'A035420', 'A003550', 'A034730'] #리스트로 종목코드 5개 지정

for num , code in enumerate(firmcode_list): #for문, firmcode_list가 참인동안
    fs_df = make_fs_dataframe(code) #make_fs_dataframe 함수를 이용해 fs_df지정
    fs_df_changed = change_df(code, fs_df) #change_df 함수를 이용해 행과 열 바꿈
    if num == 0:
        total_fs = fs_df_changed # 첫번째 행 저장
    else:
        total_fs = pd.concat([total_fs, fs_df_changed]) # 첫번째 행부터 끝까지 concat함수를 이용해 위아래로 병합.
```

```python
def make_fr_dataframe(firm_code): #재무비율 데이터 만들기함수
    fr_url = 'https://comp.fnguide.com/SVO2/asp/SVD_FinanceRatio.asp?pGB=1&cID=&MenuYn=Y&ReportGB=D&NewMenuID=104&stkGb=701&gicode=' + firm_code #fn가이드 재무비율 주소 fr_url로 지정
    fr_page = requests.get(fr_url) # fr_url 에 있는 정보를 가져와 fr_page에 저장
    fr_tables = pd.read_html(fr_page.text) #fr_page 를 판다스로 넘겨 fr_table데이터 프레임 만듬
    
    temp_df = fr_tables[0] #fr_tables[0]을 temp_df 에 저장
    temp_df = temp_df.set_index(temp_df.columns[0]) #첫번째 칼럼을 인덱스로 지정
    temp_df = temp_df.loc[['유동비율계산에 참여한 계정 펼치기',
                           '부채비율계산에 참여한 계정 펼치기',
                           '영업이익률계산에 참여한 계정 펼치기',
                           'ROA계산에 참여한 계정 펼치기',
                           'ROIC계산에 참여한 계정 펼치기']] #필요한 정보만남김
    temp_df.index = ['유동비율', '부채비율', '영업이익률', 'ROA', 'ROIC']  #temp_df의 인덱스 명을 재 설정
    return temp_df
```

```python
def make_invest_dataframe(firm_code): #투자지표 만드는 함수
    invest_url = 'https://comp.fnguide.com/SVO2/asp/SVD_Invest.asp?pGB=1&cID=&MenuYn=Y&ReportGB=D&NewMenuID=105&stkGb=701&gicode=' + firm_code #주소 지정
    invest_page = requests.get(invest_url) # 정보 page 에 저장
    invest_tables = pd.read_html(invest_page.text) #판다스로 데이터 프레임 만듬
    temp_df = invest_tables[1] #[1] 에있는 정보 temp_df에 저장
    
    temp_df = temp_df.set_index(temp_df.columns[0]) #[0]첫번째 칼럼을 인덱스로 지정
    temp_df = temp_df.loc[['PER계산에 참여한 계정 펼치기',
                           'PCR계산에 참여한 계정 펼치기',
                           'PSR계산에 참여한 계정 펼치기',
                           'PBR계산에 참여한 계정 펼치기',
                          '총현금흐름']]  #필요한 정보만 남김
    temp_df.index = ['PER', 'PCR', 'PSR', 'PBR', '총현금흐름'] #이름 새로 지정
    return temp_df
```

```python
def make_code(x):  #함수지정
    x = str(x)  #x 는 문자열
    return 'A' + '0' * (6-len(x)) + x #A + 0 + x 인데 0을 6자리 빈자리에 넣음

code_data['종목코드'] = code_data['종목코드'].apply(make_code) # apply 함수는 데이터프레임의 특정 칼럼에 주어진 함수를 적용시킨다. #종목코드칼럼에 있는 모든 거에 적용
```

```python
import time #time 모듈에는 sleep함수가 있는데 이는 코드를 잠시 멈추는 함수이다. 


for num, code in enumerate(code_data['종목코드']): #num, code 가 code_data['종목코드'] 안에 있을때
    try: #시도(1)
        print(num,code) #num, code 출력
        time.sleep(1) #1초 쉬고
        try: #시도(2)
            fs_df = make_fs_dataframe(code)  #make_fs_dataframe 함수 실행 데이터 프레임 만듬
        except requests.exceptions.Timeout: #시도(2)했다가 Timeout에러가 발행하면
            time.sleep(60) #60초간 쉬었다
            fs_df = make_fs_dataframe(code) #make_fs_dataframe 함수 실행 데이터 프레임 만듬
        fs_df_changed = change_df(code, fs_df) #행열 바꿈
        if num == 0: 
            total_fs =fs_df_changed #첫번째 행 만듬
        else:
            total_fs = pd.concat([total_fs, fs_df_changed]) #첫행부터 마지막 행까지 병합
    except ValueError: #시도(1)했다가 ValueError가 발생하면 
        continue #계속해라
    except KeyError: #시도(1)했다가 Keyerror가 발생하면 
        continue #계속해라
```
