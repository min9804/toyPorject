```python
import xlrd
```

```python
file_path = 'C:\\Users\\gjals\Desktop\\코딩\\파이썬으로 퀀트투자 만들기\\per.xlsx'
wb = xlrd.open_workbook(file_path)
```

```python
per_sh = wb.sheet_by_name('per')
```

```python
print(per_sh.row_values(1))
```

```python
per_dict = {}
for i in range(1, per_sh.nrows):
    data = per_sh.row_values(i)
    name = data[0]
    per = data[1]
    if per > 0:
        per_dict[name] = per
```

```python
import operator
```

```python
sorted_per = sorted(per_dict.items(), key=operator.itemgetter(1))
```

```python
per_rank = {}

for num, firm in enumerate(sorted_per):
    per_rank[firm[0]] = num + 1
```

```python
per_rank['삼성전자']
```

```python
roa_sh = wb.sheet_by_name('roa')
```

```python
roa_dick = {}

for i in range(1, roa_sh.nrows):
    data = roa_sh.row_values(i)
    name = data[0]
    roa = data[1]
    if roa !='':
        roa_dick[name] = roa
```

```python
sorted_roa = sorted(roa_dick.items(), key=operator.itemgetter(1), reverse=True)
```

```python
roa_rank = {}

for num, firm in enumerate(sorted_roa):
    roa_rank[firm[0]] = num + 1
```

```python
roa_rank['삼성전자']
```

```python
per_rank['삼성전자'] + roa_rank['삼성전자']
```

```python
total_rank = {}

for name in roa_rank.keys():
    if name in per_rank.keys():
        total_rank[name] = per_rank[name] + roa_rank[name]
```

```python
sorted(total_rank.items(), key=operator.itemgetter(1))
```
